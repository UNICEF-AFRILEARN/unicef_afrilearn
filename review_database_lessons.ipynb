```python
def get_youtube_ids(youtube_link):
    links=[]
    for link in youtube_link:
        if 'youtu.be' in link['videoUrl']:
            links.append(link['videoUrl'][17:])
    return links
```

```python
lessons_with_videos['youtube_id'] = lessons_with_videos['videoUrls'].apply(get_youtube_ids)
```

```python
lessons_with_videos = lessons_with_videos[lessons_with_videos['youtube_id'].str.len()!=0]
```

```python
lessons_with_videos.reset_index(inplace=True,drop=True)
```

```python
lessons_with_videos['_id'] = lessons_with_videos['_id'].astype(str)
```

```python
links_list = []
lessonid_list = []
```

```python
for index in range(len(lessons_with_videos)):
    links = lessons_with_videos.loc[index,'youtube_id']
    lessonid = lessons_with_videos.loc[index,'_id']
    for link in links:
        links_list.append(link)
        lessonid_list.append(lessonid)
```

```python
links_df = pd.DataFrame(links_list,columns=['youtube_id'])
links_df['lessonId'] = lessonid_list
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_transcript(youtubeid):
    text=""
    transcript =YouTubeTranscriptApi.get_transcript(youtubeid)
    for text_dict in transcript:
        text = text+" "+text_dict['text']
    return text
```

```python
lesson_transcript = []
for link in df['youtube_id']:
    try:
        lesson_transcript.append(get_transcript(link))
    except Exception as e:
        lesson_transcript.append("no transcript")
```

```python
df['transcripts'] = lesson_transcript
```

```python
df_with_transcripts=df[df['transcripts']!="no transcript"]
```

```python
df_with_transcripts.to_csv("lessons_with_transcripts.csv",index=[0])
```
