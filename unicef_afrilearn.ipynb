# Load recommendations

```python
school_dict = pickle.load(open("dictionary/school_level_dict.pkl",'rb'))
lesson_dict = pickle.load(open("dictionary/lesson_dict.pkl",'rb'))
```

```python
lesson_db = pd.read_csv("final_table.csv",index_col=[0])
```

```python
ldb = lesson_db[['subject_name','class_name','lessonId','title']]
```

```python
ldb= ldb.drop_duplicates()
```

```python
ldb =ldb.rename(columns={'class_name':"school_level","subject_name":"recommended_subject",'lessonId':'recommended_id'})
```

```python
def process_recommendations(df,school_level,subject,recommended_subject):
    final_df = pd.DataFrame(columns=['school_level','subject','recommended_id','recommended_subject','reason'])
    final_df['recommended_id']=df[recommended_subject]
    final_df['recommended_subject'] = recommended_subject
    final_df['subject']= subject
    final_df['school_level']= school_level
    final_df = final_df.merge(ldb, left_on=['school_level','recommended_subject','recommended_id'],right_on=['school_level','recommended_subject','recommended_id'],how='left')
    if subject==recommended_subject:
        reason_object = 'Students also watched these lessons in ' + subject
        final_df['reason'] = reason_object
    else:
        reason_object = 'Students learning '+subject+' also watched these lessons in ' + recommended_subject
        final_df['reason'] = reason_object
    return final_df
```

```python
school_levels = ldb.school_level.unique()
```

```python
%%time
full_recdf = pd.DataFrame()
for school in school_levels:
    df_schoollevel = ldb[ldb['school_level']==school]
    subjects = df_schoollevel['recommended_subject'].unique()
    for subject in subjects:
        df_lessons = df_schoollevel[df_schoollevel['recommended_subject']==subject]
        lessons = df_lessons.recommended_id.unique()
        for lesson in lessons:
            try:
                dfa= rc.recommend(school,subject,lesson)
                rec_courses = list(dfa.columns)
                rec_courses.reverse()
                dfa.columns = rec_courses
                for rec_subject in rec_courses:
                    df_rec =process_recommendations(dfa,school,subject,rec_subject)
                    if len(full_recdf==0):
                        full_recdf = df_rec.copy()
                    else:
                        full_recdf = pd.concat([full_recdf,df_rec])
            except Exception as e:
                print(school+' '+ subject +' '+ lesson)
```
