````python
## 4. Define function to get questions
def get_text(lessonid):
    content = lessons_merged[lessons_merged['lessonId']==lessonid]['content'].values[0]
    g = BeautifulSoup(content)
    contents =g.get_text().replace("\\\\n","").replace("\\\\t","").replace("\r\n","").replace("\n","")
    return contents

def get_medium_question(index):
    question = medium_questions.loc[index,'question']
    return question

def get_options(index):
    question_options = medium_questions.loc[index,'options']
    return question_options

def get_questions(prompt):
    messages = [ {"role": "system", "content": "You are a quiz generator app."} ]
    if prompt:
        messages.append(
            {"role": "user", "content": prompt},
        )
        chat = openai.ChatCompletion.create(
            model="gpt-3.5-turbo", messages=messages,temperature=0
        )
    reply = chat.choices[0].message.content
    return reply
````

## 5. run functions with prompts

````python
reply_list=[]
lessonid_list = []
reply_format =  "{'answer': 'a'}"
````

````python
prompt_format = """
         Using this text delimited in triple backticks ```{}```,
         what is the answer to this multiple choice question '{}'
         with options '{}'.Return your reply as a JSON object in this format ```{}```. 
         Do not add any notes, just return only the JSON object.
        """
````

````python
for index in range(56,len(medium_questions)):
    try:
        print(index)
        lessonid = medium_questions.loc[index,'lessonId']
        prompt =prompt_format.format(get_text(lessonid),
                                     get_medium_question(index),
                                     get_options(index),
                                     reply_format)
        reply =get_questions(prompt)
        reply_list.append(eval(reply.replace("`","")))
        lessonid_list.append(lessonid)
        df = pd.DataFrame(reply_list)
        df['lessonId'] = lessonid_list
    except Exception as e:
        print(e)
````
