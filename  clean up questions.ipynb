# 4. Get content text

```python
from bs4 import BeautifulSoup
```

```python
def get_text(content):
    g = BeautifulSoup(content)
    contents =g.get_text().replace("\\\\n","").replace("\\\\t","").replace("\r\n","").replace("\n","")
    return contents
```

# 6. TFIDF vectorizer

```python
vect = TfidfVectorizer(min_df=1, stop_words="english")                                                                                                                                                                                                   
```

```python
import numpy as np
```

```python
def cosine_sim1(text1, text2):
    tfidf = vect.fit_transform([text1, text2])
    return ((tfidf * tfidf.T).A)[0,1]
```

# 7. TFIDF for first set of questions
* To connect questions from complete_questions_firsthalf with contents from new lessons

```python
questions_df = pd.DataFrame(complete_questions_firsthalf['questions'])
```

```python
questions_df['content'] = nl['content_new']
```

```python
questions_df = questions_df.dropna()
```

```python
similarities = pd.DataFrame(index =list(questions_df['questions'].values), columns = list(questions_df['content'].values))
```

```python
for question in list(questions_df['questions'].values):
    for content in list(questions_df['content'].values):
        similarities.loc[question,content] = cosine_sim1(question, content)
```

```python
df_hard = df[df['difficulty']=='hard']
```

```python
df_hard['clean_questions'] =df_hard['questions'].apply(lambda x: str(x).replace("\nA","   A").split("   ")[0])
```

```python
df_hard['clean_options']=df_hard['questions'].apply(lambda x: str(x).replace("\nA","   A").split("   ")[1] if len(str(x).replace("\nA","   A").split("   "))>1 else "no options")
```

```python
def create_json(text):
    options = {'a':"start",'b':'start','c':'start','d':'start'}
    options['a'] = text.replace("A)","").split("B)")[0].replace("\n","").strip()
    options['b'] = text.replace("A)","").split("B)")[1].split('C)')[0].replace("\n","").strip()
    options['c'] = text.replace("A)","").split("B)")[1].split('C)')[1].split('D)')[0].replace("\n","").strip()
    options['d'] = text.replace("A)","").split("B)")[1].split('C)')[1].split('D)')[1].strip()
    return options
```

```python
df_hard = df_hard.reset_index(drop=True)
```
