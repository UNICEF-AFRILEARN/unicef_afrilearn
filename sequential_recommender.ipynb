```python
# Define the knowledge concepts and associated questions
knowledge_concepts = ['concept1', 'concept2', 'concept3']
questions = {
    'concept1': ['easy1', 'easy2'],
    'concept2': ['medium1', 'medium2'],
    'concept3': ['hard1', 'hard2']
}

# Initialize the BKT parameters
num_concepts = len(knowledge_concepts)
num_questions = sum(len(q) for q in questions.values())

# Transition matrix
L = np.eye(num_concepts)  # Learning matrix
S = np.eye(num_concepts)  # Slip matrix
G = np.eye(num_concepts)  # Guess matrix
K = np.eye(num_concepts)  # Prior knowledge matrix

# Define the learning rate, slip rate, and guess rate for each concept
learning_rates = np.array([0.3, 0.4, 0.5])
slip_rates = np.array([0.1, 0.2, 0.3])
guess_rates = np.array([0.2, 0.3, 0.4])

# Initialize the BKT parameter matrices
L = np.diag(learning_rates)
S = np.diag(slip_rates)
G = np.diag(guess_rates)

# Initialize the prior knowledge matrix
K = np.eye(num_concepts)

# Collect data (student responses)
student_responses = ['hard1']  # Example responses

for response in student_responses:
    # Find the concept associated with the question
    concept = [k for k, v in questions.items() if response in v][0]

    # Find the index of the concept and the question
    concept_idx = knowledge_concepts.index(concept)
    question_idx = questions[concept].index(response)

    # Compute the probability of success for the question
    p_success = expit(K[concept_idx, concept_idx] * G[concept_idx, concept_idx] + (1 - K[concept_idx, concept_idx]) * (1 - S[concept_idx, concept_idx]))

    # Update the prior knowledge for the concept
    K[concept_idx, concept_idx] = L[concept_idx, concept_idx] * p_success + (1 - L[concept_idx, concept_idx]) * K[concept_idx, concept_idx]

    # Recommend the next question
    next_concept_idx = np.argmax(K.diagonal())  # Choose the concept with the highest prior knowledge
    next_concept = knowledge_concepts[next_concept_idx]
    next_question = questions[next_concept][0]  # Choose the easiest question for the selected concept

    print(f"Recommended next question: {next_question}")

    # You can update the BKT parameters further based on the student's response to the recommended question
    # and repeat the process to continue recommending questions
```

```python
import numpy as np

def update_knowledge(correctness, prior_knowledge, learn, forget):
    """
    Update the knowledge state probabilities based on the correctness of a student's answer.
    
    Args:
        correctness (int): Binary value indicating the correctness of the student's answer (0 or 1).
        prior_knowledge (float): Prior probability of the student being in the mastery state.
        learn (float): Probability of transitioning from non-mastery to mastery.
        forget (float): Probability of transitioning from mastery to non-mastery.
    
    Returns:
        float: Updated probability of the student being in the mastery state.
    """
    if correctness == 1:  # Correct answer
        p_mastery = (learn * prior_knowledge) / ((learn * prior_knowledge) + ((1 - forget) * (1 - prior_knowledge)))
    else:  # Incorrect answer
        p_mastery = (prior_knowledge * (1 - learn)) / ((prior_knowledge * (1 - learn)) + ((1 - prior_knowledge) * forget))
    
    return p_mastery

# Example usage
prior_knowledge = 0.7  # Prior probability of the student being in the mastery state
learn = 0.2  # Probability of transitioning from non-mastery to mastery
forget = 0.3  # Probability of transitioning from mastery to non-mastery

# Example student responses (0: incorrect, 1: correct)
student_responses = [1, 0, 1, 1, 1]

for response in student_responses:
    prior_knowledge = update_knowledge(response, prior_knowledge, learn, forget)
    print("Updated knowledge state probability:", prior_knowledge)
```

```python
import numpy as np
from scipy.optimize import minimize

def logistic_func(theta, alpha, beta):
    """
    Logistic function used in Item Response Theory.
    
    Args:
        theta (float): Student ability parameter.
        alpha (float): Item discrimination parameter.
        beta (float): Item difficulty parameter.
    
    Returns:
        float: Probability of a correct response.
    """
    return 1 / (1 + np.exp(-(alpha * (theta - beta))))

def log_likelihood(theta, responses, items):
    """
    Log-likelihood function for Item Response Theory.
    
    Args:
        theta (float): Student ability parameter.
        responses (array-like): Array of student responses (0 or 1).
        items (array-like): Array of item parameters (alpha, beta).
    
    Returns:
        float: Log-likelihood value.
    """
    probabilities = np.array([logistic_func(theta, alpha, beta) for alpha, beta in items])
    log_likelihoods = responses * np.log(probabilities) + (1 - responses) * np.log(1 - probabilities)
    return -np.sum(log_likelihoods)

def recommend_next_question(theta, items):
    """
    Recommends the next question based on the estimated student ability parameter and item parameters.

    Args:
        theta (float): Student ability parameter.
        items (list): List of item parameters [(alpha1, beta1), (alpha2, beta2), ...].

    Returns:
        str: Recommended next question.
    """
    probabilities = np.array([logistic_func(theta, alpha, beta) for alpha, beta in items])
    next_question_index = np.argmax(probabilities)  # Select the question with the highest probability
    questions = ['easy1', 'medium1', 'hard1']  # Example questions
    return questions[next_question_index]
```

```python
# Example usage
# Simulated student responses (0: incorrect, 1: correct)
responses = [1, 0, 1]
# Item parameters (alpha: item discrimination, beta: item difficulty)
items = [(1.5, -0.5), (0.8, 0.3), (2.0, 1.0)]

# Estimate the student ability parameter using maximum likelihood estimation
initial_theta = 0  # Initial guess for the student ability
```

```python
result = minimize(log_likelihood, initial_theta, args=(np.array(responses), items), method='Nelder-Mead')
estimated_theta = result.x[0]
print("Estimated student ability parameter:", estimated_theta)
```

```python
# Recommend the next question based on the estimated student ability and item parameters
next_question = recommend_next_question(estimated_theta, items)
print("Recommended next question:", next_question)
```
